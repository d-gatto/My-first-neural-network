```python
#Importing modules and functions.
import pandas as pd
import numpy as np
from scipy.special import expit
import os
#Getting training data
os.chdir('C:\\Users\\shane\\Dropbox')
data = pd.read_csv("TrainingData.csv")
```

```python
#Initializing the neural network parameters and increment variables.
#First layer
w11 = w21 = w31 = w41 = w51 = np.array([1,1])
Dw11 = Dw21 = Dw31 = Dw41 = Dw51 = np.array([0,0])
b11 = b21 = b31 = b41 = b51 = 0
Db11 = Db21 = Db31 = Db41 = Db51 = 0
#Second layer
w12 = w22 = w32 = np.array([1,1,1,1,1])
Dw12 = Dw22 = Dw32 = np.array([0,0,0,0,0])
b12 = b22 = b32 = 0
Db12 = Db22 = Db32 = 0
#Third layer
w13 = np.array([1,1,1])
Dw13 = np.array([0,0,0])
b13 = Db13 = 0
#Defining the nodes' activation functions.
def N11(x,y):
    return expit(w11[0]*x + w11[1]*y + b11)
def N21(x,y):
    return expit(w21[0]*x + w21[1]*y + b21)
def N31(x,y):
    return expit(w31[0]*x + w31[1]*y + b31)
def N41(x,y):
    return expit(w41[0]*x + w41[0]*y + b41)
def N51(x,y):
    return expit(w51[0]*x + w51[1]*y + b51)
def N12(x,y):
    return expit(w12[0]*N11(x,y) + w12[1]*N21(x,y) + w12[2]*N31(x,y) + w12[3]*N41(x,y) + w12[4]*N51(x,y) + b12)
def N22(x,y):
    return expit(w22[0]*N11(x,y) + w22[1]*N21(x,y) + w22[2]*N31(x,y) + w22[3]*N41(x,y) + w22[4]*N51(x,y) + b22)
def N32(x,y):
    return expit(w32[0]*N11(x,y) + w32[1]*N21(x,y) + w32[2]*N31(x,y) + w32[3]*N41(x,y) + w32[4]*N51(x,y) + b32)
def N13(x,y):
    return expit(w13[0]*N12(x,y) + w13[1]*N22(x,y) + w13[2]*N32(x,y) + b13)
```

```python
#Gradient descent backwards propagation
s = .1
for k in range(100):
    for i in range(10000):
        Db13 = Db13 + data.at[i,"TruthVal"]-N13(data.at[i,"x"],data.at[i,"y"])
        Dw13 = Dw13 + Db13*np.array([N12(data.at[i,"x"],data.at[i,"y"]),N22(data.at[i,"x"],data.at[i,"y"]),N32(data.at[i,"x"],data.at[i,"y"])])
        Db12 = Db12 + Db13*w13[0]*N12(data.at[i,"x"],data.at[i,"y"])*(1-N12(data.at[i,"x"],data.at[i,"y"]))
        Dw12 = Dw12 + Db12*np.array([N11(data.at[i,"x"],data.at[i,"y"]),N21(data.at[i,"x"],data.at[i,"y"]),N31(data.at[i,"x"],data.at[i,"y"]),N41(data.at[i,"x"],data.at[i,"y"]),N51(data.at[i,"x"],data.at[i,"y"])])
        Db22 = Db22 + Db13*w13[1]*N22(data.at[i,"x"],data.at[i,"y"])*(1-N22(data.at[i,"x"],data.at[i,"y"]))
        Dw22 = Dw22 + Db22*np.array([N11(data.at[i,"x"],data.at[i,"y"]),N21(data.at[i,"x"],data.at[i,"y"]),N31(data.at[i,"x"],data.at[i,"y"]),N41(data.at[i,"x"],data.at[i,"y"]),N51(data.at[i,"x"],data.at[i,"y"])])
        Db32 = Db32 + Db13*w13[2]*N32(data.at[i,"x"],data.at[i,"y"])*(1-N32(data.at[i,"x"],data.at[i,"y"]))
        Dw32 = Dw32 + Db32*np.array([N11(data.at[i,"x"],data.at[i,"y"]),N21(data.at[i,"x"],data.at[i,"y"]),N31(data.at[i,"x"],data.at[i,"y"]),N41(data.at[i,"x"],data.at[i,"y"]),N51(data.at[i,"x"],data.at[i,"y"])])
        Db11 = Db11 + (Db12*w12[0]+Db22*w22[0]+Db32*w32[0])*N11(data.at[i,"x"],data.at[i,"y"])*(1-N11(data.at[i,"x"],data.at[i,"y"]))
        Dw11 = Dw11 + Db11*np.array([data.at[i,"x"],data.at[i,"y"]])
        Db21 = Db21 + (Db12*w12[1]+Db22*w22[1]+Db32*w32[1])*N21(data.at[i,"x"],data.at[i,"y"])*(1-N21(data.at[i,"x"],data.at[i,"y"]))
        Dw21 = Dw21 + Db21*np.array([data.at[i,"x"],data.at[i,"y"]])
        Db31 = Db31 + (Db12*w12[2]+Db22*w22[2]+Db32*w32[2])*N31(data.at[i,"x"],data.at[i,"y"])*(1-N31(data.at[i,"x"],data.at[i,"y"]))
        Dw31 = Dw31 + Db31*np.array([data.at[i,"x"],data.at[i,"y"]])
        Db41 = Db41 + (Db12*w12[3]+Db22*w22[3]+Db32*w32[3])*N41(data.at[i,"x"],data.at[i,"y"])*(1-N41(data.at[i,"x"],data.at[i,"y"]))
        Dw41 = Dw41 + Db41*np.array([data.at[i,"x"],data.at[i,"y"]])
        Db51 = Db51 + (Db12*w12[4]+Db22*w22[4]+Db32*w32[4])*N51(data.at[i,"x"],data.at[i,"y"])*(1-N51(data.at[i,"x"],data.at[i,"y"]))
        Dw51 = Dw51 + Db51*np.array([data.at[i,"x"],data.at[i,"y"]])
    w11 = w11 - s*Dw11
    b11 = b11 - s*Db11
    w21 = w21 - s*Dw21
    b21 = b21 - s*Db21
    w31 = w31 - s*Dw31
    b31 = b31 - s*Db31
    w41 = w41 - s*Dw41
    b41 = b41 - s*Db41
    w51 = w51 - s*Dw51
    b51 = b51 - s*Db51
    w12 = w12 - s*Dw12
    b12 = b12 - s*Db12
    w22 = w22 - s*Dw22
    b22 = b22 - s*Db22
    w32 = w32 - s*Dw32
    b32 = b32 - s*Db32
    w13 = w13 - s*Dw13
    b13 = b13 - s*Db13
    Dw11 = Dw21 = Dw31 = Dw41 = Dw51 = np.array([0,0])
    Db11 = Db21 = Db31 = Db41 = Db51 = 0
    Dw12 = Dw22 = Dw32 = np.array([0,0,0,0,0])
    Db12 = Db22 = Db32 = 0
    Dw13 = np.array([0,0,0])
    Db13 = 0
    print(b11,b12,b13)
```
